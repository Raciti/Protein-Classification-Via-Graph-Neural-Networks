```python
import csv
proteine = []
a = 0
b = 0
c = 0
d = 0
e = 0
f = 0
g = 0
h = 0
I = 0
j = 0
k = 0
l = 0
soglia = 299 #numero di elementi per classe da selezionare 
with open('../../Dataset/classi_proteine.csv', 'r', encoding='mac_roman') as csvfile:    
    reader = csv.reader(csvfile, delimiter = ',')
    i = 0
    for row in reader:
        if i > 3:
            
            line = row
            line = str(line[0]).split('\t')
            
            catena = line[2][0]

            if catena == 'A'  and (len(line) == 6):
                idInterno = line[1]  #aggiunta Id interno
                classe =  line[3] #aggiunta classe
                if (a <= soglia) and (classe[0] == 'a'): #1  
                    buffer = idInterno, classe
                    proteine.append(buffer)
                    a += 1
                elif (b <= soglia) and (classe[0] == 'b'): #2
                    buffer = idInterno, classe
                    proteine.append(buffer)
                    b += 1
                elif (c <= soglia) and (classe[0] == 'c'): #3  
                    buffer = idInterno, classe
                    proteine.append(buffer)
                    c += 1
                elif (d <= soglia) and (classe[0] == 'd'): #4
                    buffer = idInterno, classe
                    proteine.append(buffer)
                    d += 1
                elif (e <= soglia) and (classe[0] == 'e'): #5
                    buffer = idInterno, classe
                    proteine.append(buffer)
                    e += 1
                elif (f <= soglia) and (classe[0] == 'f'): #6
                    buffer = idInterno, classe
                    proteine.append(buffer)
                    f += 1
                elif (g <= soglia) and (classe[0] == 'g'): #7
                    buffer = idInterno, classe
                    proteine.append(buffer)
                    g += 1
                elif (h <= soglia) and (classe[0] == 'h'): #8
                    buffer = idInterno, classe
                    proteine.append(buffer)
                    h += 1
                elif (I <= soglia) and (classe[0] == 'i'): #9
                    buffer = idInterno, classe
                    proteine.append(buffer)
                    I += 1
                elif (j <= soglia) and (classe[0] == 'j'): #10
                    buffer = idInterno, classe
                    proteine.append(buffer)
                    j += 1
                elif (k <= soglia) and (classe[0] == 'k'): #11
                    buffer = idInterno, classe
                    proteine.append(buffer)
                    k += 1
                elif (l <= soglia) and (classe[0] == 'l'): #12
                    buffer = idInterno, classe
                    proteine.append(buffer)
                    l += 1    
        i+= 1
```

```python
import csv
with open('proteine.csv', mode='w') as csv_file:
    nomicolonne = ['Id', 'IdCentrale', 'Classe']
    writer = csv.DictWriter(csv_file, fieldnames=nomicolonne)
    writer.writeheader()
    for row in proteine:
        idcentrale = row[0][1:3]
        writer.writerow({'Id': row[0], 'IdCentrale': idcentrale,'Classe': row[1]})
```
